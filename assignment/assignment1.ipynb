### 2.1 Remove Useless Columns

### 2.5 Make Column 'Cluture' consistent 

```python
def clean_culture(value):
    value = value.title().strip()
    value = value.replace('Possibly ', '').replace('Probably ', '')

    replacements = {
        'Japan': 'Japanese',
        'China': 'Chinese',
        'German, Silesia (Bunzlau)': 'German',
        'German, Hanau': 'German',
        'British, Chelsea': 'British',
        'British, Stoke-On-Trent, Staffordshire': 'British',
        'Asmat People': 'Asmat',
        'Abelam People': 'Abelam'
    }
    
    if value in replacements:
        value = replacements[value]

    if 'European, Eastern' in value:
        value = 'Eastern European'
    
    return value

data['Culture'] = data['Culture'].apply(clean_culture)

for value in data['Culture'].unique():
    print(value)
```

### 2.6 Make Column 'Artist Nationality' consistent 

```python
def clean_nationality(value):
    value = value.title().strip()
    nationalities = set(value.split('|'))
    nationalities = {nat.replace('?', '').strip() for nat in nationalities}
    cleaned_value = ', '.join(sorted(nationalities))
    cleaned_value = cleaned_value.replace('Born ', '')
    return cleaned_value

data['Artist Nationality'] = data['Artist Nationality'].apply(clean_nationality)
for value in data['Artist Nationality'].unique():
    print(value)
```

### 2.7 Make Column 'Object End Date' & 'Object Begin Date' consistent

```python
invalid_rows = data[data['Object End Date'] < data['Object Begin Date']]
print("Invalid rows where End Date is smaller than Begin Date:")
invalid_rows
```

```python
data = data.drop(invalid_rows.index)
```

```python
def convert_to_bc(value):
    if value < 0:
        return f"B.C. {-value}"  # If negative, add 'B.C.' prefix and take the absolute value
    else:
        return str(value)  # Otherwise, convert the number to a string


data['Object Begin Date'] = data['Object Begin Date'].apply(convert_to_bc)
data['Object End Date'] = data['Object End Date'].apply(convert_to_bc)
data
```

### 2.8 Make Column 'Medium' consistent

```python
data['Medium'].replace({'[no medium available]': 'Unknown', 'Unknown Medium': 'Unknown'}, inplace=True)
```

### 2.9 Make Column 'Coutry' consistent

```python
country_replacements = {
    'USA': 'United States',
    'present-day Uzbekistan': 'Uzbekistan',
    'present-day France': 'France',
    'Unknown Country': 'Unknown',
}

def clean_country(country):
    return country_replacements.get(country, country)

data['Country'] = data['Country'].apply(clean_country)
data
```
